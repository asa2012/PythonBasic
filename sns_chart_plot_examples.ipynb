```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
fmri = sns.load_dataset("fmri")
ax = sns.lineplot(x="timepoint", y="signal", data=fmri)
```

```python
sns.set(style="darkgrid")
tips = sns.load_dataset("tips")

ax = sns.pointplot(x="time", y="total_bill", data=tips)
plt.show()
```

```python
ax = sns.pointplot(x="time", y="total_bill", hue="smoker",data=tips, dodge=True)
plt.show()
```

```python
ax = sns.pointplot(x="time", y="total_bill", hue="smoker",data=tips,markers=["o", "x"],linestyles=["-", "--"])
plt.show()
```

```python
ax = sns.pointplot(x="tip", y="day", data=tips)#x is continous value, y is categorical value 
plt.show()
```

```python
ax = sns.pointplot(x="tip", y="day", data=tips, join=False) 
plt.show()
```

```python
ax = sns.pointplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
plt.show()
```

```python
from numpy import median
ax = sns.pointplot(x="day", y="tip", data=tips, estimator=median)
plt.show()
```

```python
ax = sns.pointplot(x="day", y="tip", data=tips, ci=68) #show 68% cofidence interval range in chart, one standard deviation 
plt.show()
```

```python
ax = sns.pointplot(x="day", y="tip", data=tips, ci="sd")
plt.show()
```

```python
ax = sns.pointplot(x="day", y="tip", data=tips, ci=95)#show 95% confidence interval range in chart 
plt.show()
```

```python

hist_kws={'histtype': 'bar', 'edgecolor':'black', 'alpha': 0.2}
fig, ax = plt.subplots(figsize=(15, 7))
sns.distplot(tips[tips['time']=="Dinner"]['total_bill'], 
             label='dinner bill', ax=ax, hist_kws=hist_kws)
sns.distplot(tips[tips['time']=="Lunch"]['total_bill'], 
             label='lunch bill', ax=ax, hist_kws=hist_kws)
ax.set_xlabel('Bill', fontsize=12)
ax.set_ylabel('PDF', fontsize=12)#population density function
ax.legend()
plt.show()
```
